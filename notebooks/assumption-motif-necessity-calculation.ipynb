```python
from collections import Counter
```

```python
total = 10
def locations(starting, count):
    if count == 0:
        yield ()
        return
    if starting > total:
        return
    yield from locations(starting + 1, count)
    for xs in locations(starting + 2, count - 1):
        yield (starting,) + xs
```

```python
def compute_deletions_alpha(count):
    five = list(locations(0, count))
    four = list(locations(0, count - 1))
    print(f"{len(five)} configurations with {count}; {len(four)} configurations with {count}")
    p_m1 = 1 / (4 * len(five))
    p_m2 = 1 / (4 * len(four))
    print(f"P(m_1) = {p_m2 / p_m1} P(m_2)")
    counts = sorted(x for x in Counter(x[:i] + x[i+1:] for x in five for i in range(len(x))).values())
    result = sum(min(count * p_m1 / 5, p_m2) for count in counts)
    print(f"total flow for this: {result:0.3f}")
    print()
    return result
```

```python
alpha_4 = compute_deletions_alpha(4)
alpha_5 = compute_deletions_alpha(5)
alpha_6 = compute_deletions_alpha(6)
```

```python
print(f"alpha = {alpha_4 + alpha_5 + alpha_6:.3f}")
```
